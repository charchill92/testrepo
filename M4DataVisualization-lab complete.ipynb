### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# Step 1: Import necessary libraries
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Step 3: Check available tables to get the correct name (if needed)
query_tables = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
df_tables = pd.read_sql_query(query_tables, conn)
print("\nTables in the database:")
print(df_tables)

# Replace 'master' with the correct table name containing 'WorkWeekHrs', 'CodeRevHrs', and 'Age'
table_name = 'master'  # Replace with correct table if different

# Step 4: Query the 'WorkWeekHrs', 'CodeRevHrs', and 'Age' columns
try:
    query_bubble_plot = f"""
    SELECT WorkWeekHrs, CodeRevHrs, Age
    FROM {table_name}
    """
    df_bubble_plot = pd.read_sql_query(query_bubble_plot, conn)

    # Drop NaN values to focus on valid numerical data only
    df_bubble_plot = df_bubble_plot.dropna()

    # Step 5: Plot the bubble plot for 'WorkWeekHrs' and 'CodeRevHrs' with 'Age' as the bubble size
    plt.figure(figsize=(10, 6))
    plt.scatter(df_bubble_plot['WorkWeekHrs'], df_bubble_plot['CodeRevHrs'],
                s=df_bubble_plot['Age'] * 2, alpha=0.5, c='skyblue', edgecolors='k')
    plt.xlabel('Work Hours per Week')
    plt.ylabel('Code Review Hours per Week')
    plt.title('Bubble Plot of WorkWeekHrs vs. CodeRevHrs (Bubble Size: Age)')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

except Exception as e:
    print("\nError querying 'WorkWeekHrs', 'CodeRevHrs', and 'Age':", e)

# Step 6: Close the database connection
conn.close()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# Step 1: Import necessary libraries
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Step 3: Check available tables in the database
query_tables = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
df_tables = pd.read_sql_query(query_tables, conn)
print("\nTables in the database:")
print(df_tables)

# Assume the first table is the one we want, update this if needed based on the output
table_name = df_tables['Table_Name'].iloc[0]

# Step 4: Query the columns of the table to identify all column names
query_columns = f"PRAGMA table_info({table_name})"
df_columns = pd.read_sql_query(query_columns, conn)
print("\nColumns in the table:")
print(df_columns[['name']])  # Print just the column names to identify them easily

# Step 5: Find the column that likely contains database preference information
# For now, we are assuming a likely column name contains the word 'Database' or something similar

potential_columns = df_columns[df_columns['name'].str.contains('database', case=False)]['name']

if potential_columns.empty:
    print("No suitable column found for database preference information.")
else:
    # Take the first potential column that matches the criteria
    column_name = potential_columns.iloc[0]
    print(f"\nUsing column '{column_name}' for analysis.")

    # Step 6: Query the top 5 databases that respondents wish to learn next year
    query_top_databases = f"""
    SELECT {column_name}, COUNT(*) as count
    FROM {table_name}
    WHERE {column_name} IS NOT NULL
    GROUP BY {column_name}
    ORDER BY count DESC
    LIMIT 5
    """
    df_top_databases = pd.read_sql_query(query_top_databases, conn)

    # Drop NaN values to focus on valid data only
    df_top_databases = df_top_databases.dropna()

    # Step 7: Plot the pie chart for the top 5 databases
    if not df_top_databases.empty:
        plt.figure(figsize=(10, 8))
        plt.pie(df_top_databases['count'], labels=df_top_databases[column_name],
                autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightgreen', 'lightcoral', 'orange', 'violet'])
        plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
        plt.show()
    else:
        print("No data found for plotting the pie chart.")

# Step 8: Close the database connection
conn.close()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# Step 1: Import necessary libraries
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Step 3: Define the table name and column names
table_name = 'master'  # Replace with the correct table name if different
workweek_column = 'WorkWeekHrs'  # Replace with the actual column name
coderev_column = 'CodeRevHrs'    # Replace with the actual column name
age_column = 'Age'               # Replace with the actual column name

# Step 4: Query the data for age group 30 to 35
query = f"""
SELECT {workweek_column}, {coderev_column}
FROM {table_name}
WHERE {age_column} BETWEEN 30 AND 35
"""

try:
    df = pd.read_sql_query(query, conn)

    # Drop NaN values to get accurate medians
    df = df.dropna()

    # Step 5: Calculate median for 'WorkWeekHrs' and 'CodeRevHrs'
    median_workweek_hrs = df[workweek_column].median()
    median_coderev_hrs = df[coderev_column].median()

    print(f"Median WorkWeekHrs (Age 30-35): {median_workweek_hrs}")
    print(f"Median CodeRevHrs (Age 30-35): {median_coderev_hrs}")

    # Step 6: Create a stacked bar chart
    labels = ['Median Hours']
    workweek_data = [median_workweek_hrs]
    coderev_data = [median_coderev_hrs]

    plt.figure(figsize=(10, 6))
    plt.bar(labels, workweek_data, label='WorkWeekHrs', color='skyblue')
    plt.bar(labels, coderev_data, bottom=workweek_data, label='CodeRevHrs', color='lightgreen')

    plt.ylabel('Hours')
    plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
    plt.legend()
    plt.show()

except Exception as e:
    print("\nError querying the database:", e)

# Step 7: Close the database connection
conn.close()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# Step 1: Import necessary libraries
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Step 3: Define the table and column names
table_name = 'master'  # Replace with the correct table name if different
converted_comp_column = 'ConvertedComp'  # Replace with the actual column name
age_column = 'Age'  # Replace with the actual column name

# Step 4: Query the data for ages between 45 and 60
query = f"""
SELECT {age_column}, {converted_comp_column}
FROM {table_name}
WHERE {age_column} BETWEEN 45 AND 60
"""

try:
    df = pd.read_sql_query(query, conn)

    # Drop NaN values to get accurate medians
    df = df.dropna()

    # Step 5: Calculate the median 'ConvertedComp' for each age between 45 and 60
    median_comp_by_age = df.groupby(age_column)[converted_comp_column].median().reset_index()

    # Print the medians for debugging
    print("\nMedian ConvertedComp by Age (45 to 60):")
    print(median_comp_by_age)

    # Step 6: Create a line chart
    plt.figure(figsize=(10, 6))
    plt.plot(median_comp_by_age[age_column], median_comp_by_age[converted_comp_column], marker='o', linestyle='-', color='b')
    plt.xlabel('Age')
    plt.ylabel('Median Compensation')
    plt.title('Median ConvertedComp for Ages 45 to 60')
    plt.grid(True)
    plt.show()

except Exception as e:
    print("\nError querying the database:", e)

# Step 7: Close the database connection
conn.close()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# Step 1: Import necessary libraries
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Step 3: Define the table and column names
table_name = 'master'  # Replace with the correct table name if different
main_branch_column = 'MainBranch'  # Replace with the actual column name

# Step 4: Query the data from the MainBranch column
query = f"""
SELECT {main_branch_column}
FROM {table_name}
"""

try:
    df = pd.read_sql_query(query, conn)

    # Drop NaN values
    df = df.dropna()

    # Step 5: Count the occurrences of each value in MainBranch
    branch_counts = df[main_branch_column].value_counts()

    # Print the counts for debugging
    print("\nCounts of MainBranch values:")
    print(branch_counts)

    # Step 6: Create a horizontal bar chart
    plt.figure(figsize=(10, 6))
    plt.barh(branch_counts.index, branch_counts.values, color='skyblue')
    plt.xlabel('Count')
    plt.ylabel('MainBranch')
    plt.title('Distribution of MainBranch')
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.show()

except Exception as e:
    print("\nError querying the database:", e)




```
